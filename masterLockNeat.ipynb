```python
def getASIN(azQuery, manuPartNumber):   
    query = sp_api.api.Catalog(credentials=credentials).list_items(MarketplaceId='ATVPDKIKX0DER', Query=azQuery)
    skuMatch = False
    if len(query.Items) >0:
        for i in query.Items:
            #if manufact sku matches part num found on amazon
            if 'PartNumber' in i['AttributeSets'][0].keys() and manuPartNumber == i['AttributeSets'][0]['PartNumber']:

                skuMatch = True
                #the match id returned
                match='True'
                asin = i['Identifiers']['MarketplaceASIN']['ASIN']

                if len(i['SalesRankings']) > 0:
                    salesRankId = i['SalesRankings'][0]['ProductCategoryId']
                    salesRank = i['SalesRankings'][0]['Rank']
                else:
                    salesRankId= 'No Rank'
                    salesRank= 'No Rank'                            

                listPartNum = i['AttributeSets'][0]['PartNumber']

                if 'PackageDimensions' in i['AttributeSets'][0].keys() and 'Height' and 'Width' and 'Length' in i['AttributeSets'][0]['PackageDimensions']:
                    boxV = round(i['AttributeSets'][0]['PackageDimensions']['Height']['value'] \
                    * i['AttributeSets'][0]['PackageDimensions']['Length']['value'] \
                    * i['AttributeSets'][0]['PackageDimensions']['Width']['value'], 2)
                    boxU = i['AttributeSets'][0]['PackageDimensions']['Width']['Units']

                    boxVolume = f'{boxV} {boxU}'
                else:

                    boxVolume = 'None'


                break

        if not skuMatch:
            #defaults to first listing  
            #if amazon has part number listed but it does not match sku
            if 'PartNumber' in query.Items[0]['AttributeSets'][0].keys():

                match='False'


                asin = query.Items[0]['Identifiers']['MarketplaceASIN']['ASIN']

                if len(query.Items[0]['SalesRankings']) > 0:
                    salesRankId = query.Items[0]['SalesRankings'][0]['ProductCategoryId']
                    salesRank = query.Items[0]['SalesRankings'][0]['Rank']
                else: 
                    salesRankId= 'No Rank'
                    salesRank= 'No Rank'            

                listPartNum = query.Items[0]['AttributeSets'][0]['PartNumber']

                if 'PackageDimensions' in query.Items[0]['AttributeSets'][0].keys() and 'Height' and 'Width' and 'Length' in query.Items[0]['AttributeSets'][0]['PackageDimensions']:
                    boxV = round(query.Items[0]['AttributeSets'][0]['PackageDimensions']['Height']['value'] \
                    * query.Items[0]['AttributeSets'][0]['PackageDimensions']['Length']['value'] \
                    * query.Items[0]['AttributeSets'][0]['PackageDimensions']['Width']['value'], 2)
                    boxU = query.Items[0]['AttributeSets'][0]['PackageDimensions']['Width']['Units']

                    boxVolume = f'{boxV} {boxU}'
                else:

                    boxVolume = 'None'

            #if amazon listing does not have partNumber listed
            else:

                match='Not Listed'
                asin = query.Items[0]['Identifiers']['MarketplaceASIN']['ASIN']

                if len(query.Items[0]['SalesRankings']) > 0:
                    salesRankId = query.Items[0]['SalesRankings'][0]['ProductCategoryId']
                    salesRank = query.Items[0]['SalesRankings'][0]['Rank']
                else: 
                    salesRankId= 'No Rank'
                    salesRank= 'No Rank'


                listPartNum = 'None'

                #can put box dims here
                boxVolume = 'None'

        print(match)
    
    #query error
    else:
        asin = 'None'
        salesRankId = 'None'
        salesRank= 'None'
        boxVolume = 'None'
        listPartNum = 'None'
        match = 'Empty'
            
    
    return asin, salesRankId, salesRank, boxVolume, listPartNum, match
```

**Get Market Price from Products api**

```python
def getMarketPrice(asin):
    query = sp_api.api.Products(credentials=credentials).get_item_offers(asin=asin, item_condition=["New"],
                                                                         MarketplaceId ='ATVPDKIKX0DER')
    #checks if there are any sellers offering the product on amzn
    if len(query.Offers) > 0:
        
        bbw = [x for x in query.Offers if x['IsBuyBoxWinner'] == True]
        if not bbw:
            #if there is not a buybox winner
            isBBW = 'False'
            marketPrice = min([x['ListingPrice']['Amount'] for x in query.Offers])

        else:
            #if there is  a buybox winner
            isBBW = 'True'
            marketPrice = bbw[0]['ListingPrice']['Amount']
            
    else:
        isBBW = 'False'
        marketPrice = 'no offers'

    return isBBW, marketPrice
```

**Query Fee API for FBA calculator**

```python
def getFees(asin, marketPrice):
    query = sp_api.api.ProductFees(credentials=credentials).get_product_fees_estimate_for_asin(asin=asin,price=marketPrice,
                                                                                               currency='USD',is_fba=True,
                                                                                               MarketplaceId='ATVPDKIKX0DER') 
    if query.payload['FeesEstimateResult']['Status'] == 'Success':
        
        #must add 99 (1) cent fixed closing fees
        fees = float(query.payload['FeesEstimateResult']['FeesEstimate']['TotalFeesEstimate']['Amount']) + 1
        
    else: fees = 'No FBA'
    
    return fees
```
